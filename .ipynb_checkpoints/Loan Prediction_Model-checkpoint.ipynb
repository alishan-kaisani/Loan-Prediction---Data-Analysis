```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.svm import KNeighborsClassifier
import math
```

```python
df = pd.read_csv("Loan Prediction Test - final.csv")
df = df.drop("Unnamed: 0",axis=1)
Loan_ID_test = df.pop("Loan_ID")
```

```python
print(df.columns.values)
df.head()
```

```python
train_df = pd.read_csv("Loan Prediction Train - final.csv")
train_df = train_df.drop("Unnamed: 0",axis = 1)
Loan_ID_train = train_df.pop("Loan_ID")
```

```python
train_df.head()
```

### Train Model

```python
Y = train_df.pop("Loan_Status")
X = train_df
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
# model = LogisticRegression(random_state=100)
model = RandomForestClassifier(n_estimators=30,max_depth=6,min_samples_leaf=4,oob_score=True,random_state=100)
model = RFE(model,18)
model.fit(X_train,y_train)
```

### Most Important Features:

```python
d = {}
for i in range(0,len(df.columns.values[model.support_])):
    d[df.columns.values[model.support_][i]] = round(model.estimator_.feature_importances_[i],3)
print(d)
sorted(d,key=d.get)
```

### Score against training subset

```python
print("Model's mean accuracy is: {}".format(model.score(X_test,y_test)))
```

### Predict on Actual Test Data

```python
X_real = df
y_pred = model.predict(X_real)
y_pred = pd.Series(model.predict(X_real))
```

```python
final_df = pd.DataFrame(Loan_ID_test)
final_df["Loan_Status"] = y_pred
final_df["Loan_Status"] = final_df["Loan_Status"].map({1:"Y",0:"N"})
final_df.head()
```
